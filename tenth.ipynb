```python
import pandas as pd

feats = pd.read_csv(
    'Applied-Deep-Learning-with-Keras/Lesson02/bank_data_feats.csv',
    index_col = 0
)
target = pd.read_csv(
    'Applied-Deep-Learning-with-Keras/Lesson02/bank_data_target.csv',
    index_col = 0
)
```

```python
from sklearn.model_selection import train_test_split

test_size = 0.2
random_state = 42

X_train, X_test, y_train, y_test = train_test_split(
    feats, target, test_size=test_size, random_state=random_state
)

from keras.models import Sequential

model = Sequential()

from keras.layers import Dense, Activation

input_shape = X_train.shape[1]
units = 1

model.add(Dense(units, input_dim = input_shape))
model.add(Activation('tanh'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x=X_train, 
    y=y_train['y'],
    epochs=20, 
    batch_size=32, 
    validation_split=0.2
)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Loop')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Loop')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
# Euler Formula
# e^ix = cos(x) + i sin(x)
# e^-ix = cons(x) - i sin(x)
# e^ix + e^-ix = 2cos(x)
# cos(x) = (e^ix + e^-ix) / 2

# e^ix - e^-ix = 2i sin(x)
# sin(x) = (e^ix - e^-ix) / 2
# tan(x) = sing(x) / cos(x)

# cosh(x) = (e^x + e^-x) / 2
# sinh(x) = (e^x - e^-x) / 2
# tanh(x) = sinh(x) / cosh(x) >> -1 ~ 1 사이의 값을 갖는다.
```

```python
# x: 1 by 2
# W1: 2 by 3
# b1: 1 by 3
# 행렬 곱셉의 규칙 >> (1 by 2)*(2 by 3)+(1 by 3)

# a1: (1 by 3)
# w2: 3 by 5
# b2: 1 by 5
# z2: (1 by 3)*(3 by 5)+(1 by 5) >> (1 by 5)

# a2: (1 by 5)
# W3: (5 by 1)
# b3: (1 by 1)
# z3 = a2 * W3 + b3 >> (1 by 5)(5 by 1)+(1 by 1) >> (1 by 1)

# Y = sigmoid(z3) >> 0 ~ 1 확률로 나타남
```

```python
import numpy as np
from sklearn.datasets import make_classification

X, y = make_classification(
    # 데이터 300개
    n_samples=300,
    
    # 독립변수 10개
    n_features=10, 
    
    # 나올 수 있는 결과 2개
    n_classes=2
)

print("Num of Dataset:", X.shape[0])
print("Num of Features:", X.shape[1])
print("Possible Output Classes", np.unique(y))
```

```python
# Features가 10개라는 뜻은 
# y = ax1 + bx2 + cx3 + dx4 + ex5 fx6 gx7 hx8 ix9 jx10 + alpha
```

```python
from keras.models import Sequential
model = Sequential()

from keras.layers import Dense, Activation

# Dense(출력 개수, 활성화 함수, 입력 개수)
model.add(Dense(20, activation='tanh', input_dim=10))
model.add(Dense(40, activation='tanh'))
model.add(Dense(5, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='sgd', loss='binary_crossentropy')
```

```python
# model.fit(X, y, epochs=100, batch_size=5 verbose=0)
# verbose=0은 출력을 하지 않음
history = model.fit(X, y, epochs=1000, batch_size=5)
```

```python
print(history.history)
```

```python
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Loop')
plt.show()
```
