```python
def plot_data(pl, X, y):
    # y가 0인 애들 중에 x좌표 X[y == 0,0]
    # y가 0인 애들 중에 y좌표 X[y == 0,1]
    pl.plot(X[y == 0,0], X[y == 0, 1], 'ob', alpha = 0.5)
    
    # y가 1인 애들 중에 x좌표 X[y == 0,0]
    # y가 1인 애들 중에 y좌표 X[y == 0,1]
    pl.plot(X[y == 1,0], X[y == 1, 1], 'xr', alpha = 0.5)
    
    # 범례 자동 지정
    pl.legend(['0', '1'])
    return pl
```

```python
def plot_decision_bound(model, X, y):
    # X shape -> (400, 2)
    # a = x, b = y
    amin, bmin = X.min(axis = 0) - 0.1
    amax, bmax = X.max(axis = 0) + 0.1
    #print("amin =", amin)
    #print("bmin =", bmin)
    #print("amax =", amax)
    #print("bmax =", bmax)
    
    # np.linspace()는
    # 첫번째 인자 ~ 두번째 인자 사이의 값을 세번째 인자의 개수만큼 균일하게 쪼갠다.
    hticks = np.linspace(amin, amax, 101)
    vticks = np.linspace(bmin, bmax, 101)
    
    # meshgrid()는 x축 정보(벡터), y축 정보(벡터)를 입력 받아 사각형 영역에 대한 정보를 반환한다.
    # 실제 데이터를 전부 포함할 수 있는 컴팩트한 사각형의 범위를 잡는다.
    aa, bb = np.meshgrid(hticks, vticks)
    
    # np.c: 두 개의 1차원 배열을 칼럼으로 세로로 붙여서 2차원 배열 만들기
    # ravel: 배열의 shape을 1차원 배열로 만드는 메서드
    ab = np.c_[aa.ravel(), bb.ravel()]
    
    # model.predict(ab)를 사용하여 아래쪽에서 contourf를 사용할 준비를 한다.
    # c라는 정보는 결국 Contour에서 활용하는 등고선의 기준 높이로 활용된다.
    c = model.predict(ab)
    
    #print(c)
    Z = c.reshape(aa.shape)
    
    # cmap, alpha는 **kwargs이 처리하게 되고 aa, bb, Z는 *args이 처리하게 된다.
    # cmap == ColorMap을 bwr을 택할 경우 아래쪽은 파랑색, 가운데는 흰색, 위쪽은 빨간색
    # alpha값이란 것은 투명도를 의미한다. alpha값은 낮을수록 투명가 높으며 값이 높은 경우엔 색상이 진하게 칠해진다.
    # plt.contourf를 통해서 등고선이 같은애들의 범주를 잡아줬다.

    plt.contourf(aa, bb, Z, cmap = 'bwr', alpha = 0.2)
    
    # plot_data를 통해서 실제 데이터들을 그래프상에 그렸다.
    plot_data(plt, X, y)
    return plt
```

```python
import pandas as pd

data = pd.read_csv(
    "./Applied-Deep-Learning-with-Keras/Lesson02/bank_data_feats.csv",
    index_col = 0
)
y = pd.read_csv(
    "./Applied-Deep-Learning-with-Keras/Lesson02/bank_data_target.csv",
    index_col = 0
)

X = data
```

```python
print("Num of Dataset = ", data.shape[0])
print("Num of Feature = ", data.shape[1])
print(y.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

def build_model():
    model = Sequential()
    
    model.add(Dense(16, input_dim=X.shape[1], activation = 'relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model
```

```python
import numpy as np

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
# 어떤 특정한 시작 숫자를 정해 주면 컴퓨터가 정해진 알고리즘에 의해 
# 마치 난수처럼 보이는 수열을 생성한다. 이런 시작 숫자를 시드(seed)라고 한다.
seed = 1
np.random.seed(seed)
```

```python
n_folds = 5
epochs = 300
batch_size = 5

# 분류 작업을 할 때 사용하는 분류기 생성자
classifier = KerasClassifier(
    # 분류를 수행하는데 실질적으로 사용할 모델을 만드는 함수
    build_fn = build_model,
    
    # 몇 회를 반복할 것인가
    epochs = epochs,
    batch_size = batch_size,
    verbose = 1
)
```

```python
# 우리가 가지고 있는 Train, Test 데이터에 대한 검증 생성자
# Train, Test 총합 1000개가 있다고 가정
# 1. Train 0 ~ 799, Test 800 ~ 999
# 2. Train 200 ~ 999, Test 0 ~ 199
# 3. Train 0 ~ 199, 400 ~ 999 Test 200 ~ 399
# .........
kfold = StratifiedKFold(
    # 전체 구간을 5번으로 분할해서 검증하도록 만든다.
    n_splits=n_folds,
    shuffle=True,
    # 다음 모델에서도 동일한 fitting을 사용해서 한다.
    random_state=seed
)
```

```python
# 위에서 만든 분류기와 검증기를 전달해서 x, y 값에 대한 교차 검증을 수행한다.
result = cross_val_score(classifier, X, y, cv=kfold)
```

```python
for i in range(n_folds):
    print('Accuracy at fold', i+1, '=', result[i])
    
print("")
print("Final Cross Valid Test Accuracy", result.mean())
print("Standard Deviation of Filnal Test Acuuracy", result.std())
```

```python
def model_build1(activation='relu', optimizer='adam'):
    model = Sequential()
    
    model.add(Dense(8, input_dim=X.shape[1], activation=activation))
    model.add(Dense(8, input_dim=8, activation=activation))    
    model.add(Dense(8, input_dim=8, activation=activation))    
    model.add(Dense(1, input_dim=8, activation='sigmoid'))    
    
    model.compile(
        loss = 'binary_crossentropy',
        optimizer = optimizer,
        metrics=['accuracy']
    )
    
    return model

def model_build2(activation='relu', optimizer='adam'):
    model = Sequential()
    
    model.add(Dense(16, input_dim=X.shape[1], activation=activation))
    model.add(Dense(8, input_dim=8, activation=activation))    
    model.add(Dense(4, input_dim=8, activation=activation))    
    model.add(Dense(1, input_dim=8, activation='sigmoid'))    
    
    model.compile(
        loss = 'binary_crossentropy',
        optimizer = optimizer,
        metrics=['accuracy']
    )
    
    return model

def model_build3(activation='relu', optimizer='adam'):
    model = Sequential()
    
    model.add(Dense(8, input_dim=X.shape[1], activation=activation))
    model.add(Dense(4, input_dim=8, activation=activation))    
    model.add(Dense(1, input_dim=8, activation='sigmoid'))    
    
    model.compile(
        loss = 'binary_crossentropy',
        optimizer = optimizer,
        metrics=['accuracy']
    )
    
    return model
```

```python
n_folds = 5
batch_size = 5
epochs = 200

results = []

models = [model_build1, model_build2, model_build3]

```

```python
# 인공 신경망 모델에 대한 교차 검증하여 어떤 모델이 가장 성능이 좋은 지를 판정
for idx in range(len(models)):
    classifier = KerasClassifier(
        build_fn = models[idx],
        epochs = epochs,
        batch_size = batch_size,
        verbose = 1
    )
    
    kfold = StratifiedKFold(
        n_splits = n_folds,
        shuffle = True,
        random_state = seed
    )
    
    result = cross_val_score(classifier, X, y, cv = kfold)
    results.append(result)
```

```python
for idx in range(len(models)):
    print('Model', idx+1, "Test Accuracy =", results[idx].mean())
```

```python
# Hyper Parameter(초 매개변수)
# 교차 검증을 통해서 어떤 녀석을 활용하는 것이 좋은지 판정할 수 있다.
optimizers = ['rmsprop', 'adam', 'sgd']
activations = ['relu', 'tanh']

for opt in range(len(optimizers)):
    for active in range(len(activations)):
        optimizer = optimizers[opt]
        activation = activations[active]
        
        classifier = KerasClassifier(
            build_fn = model_build1,
            epochs = 300,
            batch_size = 5,
            verbose = 1
        )
        
        kfold = StratifiedKFold(
            n_splits = n_folds,
            # 카드를 섞는 기능
            shuffle = True,
            random_state = seed
        )
        
        result = cross_val_score(classifier, X, y, cv = kfold)
        
        results.append(result)
```
