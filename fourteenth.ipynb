```python
# pd.get_dummies() >> 카테고리화 할 수 있는 변수들을 변환, 빈도수 체크
SpecType = pd.get_dummies(
    X.iloc[:, 1], 
    
    # 가변수의 첫번째 변수를 자동으로 삭제
    drop_first = True,
    prefix = 'SpecType'
)
```

```python
preDie = pd.get_dummies(
    X.iloc[:, 2], 
    drop_first = True,
    prefix = 'preDie'
)
```

```python
print(preDie)
```

```python
X.drop(
    [
        'Admission_type', 
        'PreExistingDisease', 
        'Gender'
    ],
    axis = 1,
    inplace = True
)
```

```python
print(X)
```

```python
X = pd.concat(
    [X,SpecType, preDie], axis=1
)
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, 
    
    # 재현율
    random_state = 1
)
```

```python
# StandardScaler: 평균을 제거하고 데이터를 단위 분산으로 조정
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
 
# Z스코어 베이스의 스케일링
X_train = sc.fit_transform(X_train)
print(X_train)
```

```python
X_train = pd.DataFrame(
    X_train, 
    
    # colunms은 X_test의 것을 사용
    columns = X_test.columns
)

print(X_train)
```

```python
print(X_test.shape)

X_train.head()
```

```python
X_test = sc.transform(X_test)
X_test = pd.DataFrame(
    X_test,
    columns=X_train.columns
)
```

```python
X_train_np = X_train.values
y_train_np = y_train.values
X_test_np = X_test.values
y_test_np = y_test.values
```

```python
print(type(X_train_np))

# 데이터가 10만개 이하일 경우 Numpy가 느려짐
# Numpy는 10만개 이상에서 사용하는 것이 효과적
# Numpy는 c++기반, 수치해석을 위한 라이브러리
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
```

```python
model = Sequential()

model.add(Dense(units=6, activation='relu', kernel_initializer='uniform', input_dim=10))
model.add(Dropout(rate=0.3))
```

```python
model.add(Dense(units=6, activation='tanh', kernel_initializer='uniform'))
```

```python
model.add(Dense(units=6, activation='relu', kernel_initializer='uniform'))
model.add(Dropout(rate=0.3))
```

```python
model.add(Dense(units=6, activation='tanh', kernel_initializer='uniform'))
```

```python
model.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(
    X_train_np, y_train_np, 
    epochs=400, 
    
    # 계산을 할 떄 몇 개의단위로 할 것인가
    batch_size=20
)
```

```python
y_pred_class = model.predict(X_test_np)
y_pred_prob = model.predict_proba(X_test_np)
```

```python
from sklearn.metrics import roc_curve
```

```python
# frp(거짓 양성률): 실제론 해당사항이 없지만, 해당 사항이 있다고 판정을 내리는 경우
# tpr()
# threadholds = 임계치
fpr, tpr, threadholds = roc_curve(y_test, y_pred_prob)
```

```python
print(fpr)
```

```python
print(tpr)
```

```python
print(threadholds)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fpr, tpr)
plt.title("ROC Curve")
plt.xlabel("False Positive rate")
plt.ylabel("True Positive rate")
plt.grid(True)
plt.show()
```

```python
def optimum_threshold(my_threshold):
    # 참인 것을 참이라고 할 확률
    print("Sensitivity: ", tpr[threadholds > my_threshold][-1])
    
    # 아닌걸 아니라고 할 확률
    print("Specificity: ", 1 - fpr[threadholds > my_threshold][-1])    
```

```python
print(tpr[threadholds > 0.5])
```

```python
# 맨 뒤에 배열 원소를 검색
print(tpr[threadholds > 0.5][-1])
```
